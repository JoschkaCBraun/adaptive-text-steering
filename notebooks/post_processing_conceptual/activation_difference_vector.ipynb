```python
import matplotlib.pyplot as plt
import numpy as np

# Generate samples from two 2D Gaussian distributions
n_samples = 10
pos_mean = np.array([1.5, 0.5])
neg_mean = np.array([-1.5, -0.5])
cov = np.array([[0.2, 0], [0, 0.2]])

positive_points = np.random.multivariate_normal(pos_mean, cov, n_samples)
negative_points = np.random.multivariate_normal(neg_mean, cov, n_samples)

# Calculate actual means of sampled points
pos_mean_actual = np.mean(positive_points, axis=0)
neg_mean_actual = np.mean(negative_points, axis=0)

# Create plot
plt.figure(figsize=(7, 4))
plt.axhline(0, color='black', linewidth=1, zorder=1)
plt.axvline(0, color='black', linewidth=1, zorder=1)

# Plot points
plt.scatter(positive_points[:, 0], positive_points[:, 1], color='red', s=50, 
           label='Positive Activations', zorder=3)
plt.scatter(negative_points[:, 0], negative_points[:, 1], color='blue', s=50, 
           label='Negative Activations', zorder=3)

# Plot means
plt.scatter(pos_mean_actual[0], pos_mean_actual[1], color='red', marker='*', s=200,
           label='Positive Mean', zorder=4, edgecolor='black')
plt.scatter(neg_mean_actual[0], neg_mean_actual[1], color='blue', marker='*', s=200,
           label='Negative Mean', zorder=4, edgecolor='black')

# Draw arrows
for i in range(len(positive_points)):
    plt.arrow(negative_points[i, 0], negative_points[i, 1],
             positive_points[i, 0] - negative_points[i, 0],
             positive_points[i, 1] - negative_points[i, 1],
             head_width=0.15, head_length=0.2, fc='grey', ec='grey',
             length_includes_head=True, alpha=0.5, zorder=2)

# Add arrow to legend
plt.plot([], [], color='grey', alpha=0.5, label='Activations Difference Vector',
         marker='>', markersize=10, linestyle='-', linewidth=1)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Activation Differences')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
